```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('main_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.loc[:,['STATE', 'DISTRICT', 'BLOCK_NAME', 'LAT', 'LON', 'YEAR_OBS', 'MONSOON', 'POMRB', 'POMKH', 'PREMON']]
```

```python
df
```

```python
df = pd.melt(df, id_vars=['STATE', 'DISTRICT', 'BLOCK_NAME', 'LAT', 'LON', 'YEAR_OBS'], value_vars=['MONSOON', 'POMRB', 'POMKH', 'PREMON'], value_name='Depth', var_name='Season')
```

```python
df
```

```python
state2dist = dict()
state_list = set(df.STATE)
for state in state_list:
    state2dist[state] = list(set(df[df.STATE == state].DISTRICT))
state2dist
```

```python
state_list = ['AP', 'AR', 'AS', 'BR','CG', 'CH', 'DD', 'DL', 'GA', 'GJ', 'HP', 'HR', 'JH', 'JK', 'KA', 'KL', 'ML', 'MN', 'MP', 'MS', 'NL', 'OD', 'PB', 'PY', 'RJ', 'TG', 'TN', 'TR', 'UK', 'UP', 'WB']
```

```python
%time
data = pd.DataFrame()
for state in state_list:
    districtData = pd.DataFrame()
    temp = df[df.STATE == state]
    districtNameList = state2dist[state]
    for district in districtNameList:
        try:
            singleDistrictData = pd.DataFrame()
            temp = df[df.DISTRICT == district]
            districtCsvData = pd.read_csv(f'../BLOCK-ALGO-DONE/CSV - Copy/BLOCKWISE_{state}_{district}.csv')
            blockNameList = []
            for col in districtCsvData.columns:
                if col not in ['Date', 'Season', '_', 'Unnamed'] and '_' not in col:
                    blockNameList.append(col)
            for block in blockNameList:
                singleBlockData = pd.DataFrame()
                temp = df[df.BLOCK_NAME == block]
                for i in range(1996,2019):
                    temp = df[df.YEAR_OBS == i]
                    singleBlockYearData = pd.DataFrame()
                    singleBlockYearData['STATE']=[state for j in range(4)]
                    singleBlockYearData['DISTRICT']=[district for j in range(4)]
                    singleBlockYearData['BLOCK_NAME']=[block for j in range(4)]
                    try:
                        lat = sum(set(temp.LAT))/len(set(temp.LAT))
                        lon = sum(set(temp.LON))/len(set(temp.LON))
                    except:
                        lat = np.nan
                        lon = np.nan
                    singleBlockYearData['LAT'] = [lat for j in range(4)]
                    singleBlockYearData['LON'] = [lon for j in range(4)]
                    singleBlockYearData['YEAR_OBS'] = [i for j in range(4)]
                    singleBlockYearData['Season'] = ['POMRB', 'PREMON', 'MONSOON', 'POMKH']
                    singleBlockData = singleBlockData.append(singleBlockYearData)
                singleBlockData['Depth'] = districtCsvData[block]
                for algo in ['Trend', 'SMA', 'EWMA', 'DES', 'TES', 'ARIMA']:
                    singleBlockData[f'Depth_{algo}'] = list(districtCsvData[f'{block}_{algo}'])
                singleDistrictData = singleDistrictData.append(singleBlockData)
            districtData = districtData.append(singleDistrictData)
        except:
            pass
    data = data.append(districtData)
```

```python
data.to_csv('KeplerData.csv')
```
